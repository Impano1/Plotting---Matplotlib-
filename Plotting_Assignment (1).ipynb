```python
# YOUR CODE HERE
# extracting the co2 datas
co2_data = fuel_econ_data.co2
# skewing for the co2 data
skew = stats.skew(co2_data)
print(skew)
```

```python
# YOUR CODE HERE
bin_width = 20
# Set up the plot parameters and display the histogram
plt.figure(figsize = [8,3])
plt.hist(co2_data, bins = bin_width, color = 'brown', alpha = 0.7)
plt.xlabel('CO2 (g/km)')
plt.ylabel('Frequency')
plt.title('CO2 Emissions')
plt.savefig('co2 Emissions')
```

```python
# YOUR CODE HERE
fuel_data = fuel_econ_data.fuelType
bin_width = 15
plt.hist(fuel_data, bins=bin_width, color='green')
plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.title('Distribution of Fuel Types')
plt.show()
plt.savefig('distribution of Fuel Types')
```

```python
high_way_data = fuel_econ_data.UHighway
bin_width = 15
plt.hist(high_way_data, bins = bin_width, color = 'pink')
plt.xlabel(' Fuel Economy')
plt.ylabel('Frequency')
plt.title('Highway Data Distribution')
plt.savefig('Highway Data Distribution')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


columns = ['id', 'year', 'cylinders', 'displ', 'pv2', 'pv4', 'city', 'UCity', 'highway', 'UHighway', 'comb', 'co2', 'feScore', 'ghgScore']


correlation = fuel_econ_data[columns].corr()


sns.heatmap(correlation, annot=False, cmap='RdBu', square=True)
plt.title('Heat Map of Correlation Data')
plt.xticks(rotation=45, ha='right')
plt.show()
plt.savefig('Heat Map of Correlation Data')
```
